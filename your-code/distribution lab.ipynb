## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""

total_fruits = 100
apples = 60
oranges = 40

P_apple = apples / total_fruits
P_orange = oranges / total_fruits

P_apple_rounded = round(P_apple, 3)
P_orange_rounded = round(P_orange, 3)

# Display
print(f"Probability of picking an apple: {P_apple_rounded}")
print(f"Probability of picking an orange: {P_orange_rounded}")
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
# What is the probability that the first 5 fruits are all apples?
first5apples = P_apple ** 5
first5applesthen15orange = P_apple ** 5 * P_orange ** 15

print(f"Probability that the first 5 fruits are all apples: {round(first5apples, 3)}")
print(f"Probability that the first 5 fruits are all apples and the next 15 fruits are all oranges: {round(first5applesthen15orange, 3)}")
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
# your solution here
import math
n = 20 # sample amount
k = 5 # apples
binomialco = math.comb(n, k)

solution = binomialco * (P_apple ** k) * (P_orange ** (n - k))

print( f'The probability of getting exactly 5 apples and 15 oranges is {round(solution, 3)}')
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python

lessthan5 = sum(math.comb(n, k) * (P_apple ** k) * (P_orange ** (n - k)) for k in range(5))
print(f'The probability that less than 5 fruits picked are apples is {round(lessthan5, 4)}')
````

**Plot the PDF (probability distributiom function) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
x_values = list(range(0, n + 1))  # Possible values for the number of apples
y_values = [math.comb(n, k) * (P_apple ** k) * (P_orange ** (n - k)) for k in x_values]

plt.figure(figsize=(10, 6))
plt.bar(x_values, y_values, color='skyblue', edgecolor='black')
plt.xlabel('Number of Apples (k)', fontsize=12)
plt.ylabel('Probability P(X = k)', fontsize=12)
plt.title('Binomial Distribution PDF (20 Fruits, P(Apple) = 0.60)', fontsize=14)
plt.xticks(range(0, n+1, 1))
plt.grid(True)
plt.show()
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
import math


lambda_ = 2.3 # average number of goals
k = 5 # number of goals

# Poisson probability mass function
pkgoals = (lambda_**k * math.exp(-lambda_)) / math.factorial(k)


print(f"The probability of scoring exactly {k} goals is: {round(pkgoals, 6)}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python

k_values = list(range(0, 11)) 

# Poisson probability mass function for each k
poisson_probs = [(lambda_**k * math.exp(-lambda_)) / math.factorial(k) for k in k_values]

plt.figure(figsize=(10, 6))
plt.bar(k_values, poisson_probs, color='skyblue', edgecolor='black')
plt.xlabel('Number of Goals (k)', fontsize=12)
plt.ylabel('Probability P(X = k)', fontsize=12)
plt.title('Poisson Distribution for Number of Goals (λ = 2.3)', fontsize=14)
plt.xticks(k_values)
plt.grid(True)
plt.show()
````
